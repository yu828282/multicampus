```python
class Movie:
    def __init__(self): # 연결
        self.con = sqlite3.connect("tutorial.db", timeout=5, isolation_level=None)
        self.cur = self.con.cursor()
        
    def __del__(self):
        self.cur.close()
        self.con.close()# 연결해제
        
    def movie_insert(self, no, title, year, score):
        sql = "INSERT INTO movie VALUES (?,?,?,?)"
        self.con.execute(sql,(no, title, year, score))
        self.con.commit()
        
    def movie_update(self, title, year, score, no):       
        sql = "UPDATE movie SET title = ?, year= ?, score =?  where no =?"
        self.con.execute(sql,(title, year, score, no,))
        self.con.commit()
        
    def movie_delete(self, no):             
        sql = "DELETE FROM movie where no =?"
        self.con.execute(sql, (no,))
        self.con.commit()           
    
    def movie_selectAll(self):            
        sql = "SELECT * FROM movie"
        res = self.cur.execute(sql,)
        for _ in res:            
            print(_)
        return res
        
    def movie_selectOne(self, no):      
        sql = "SELECT * FROM movie Where  no = ?"
        self.cur.execute(sql, (no,))
        row = self.cur.fetchone()
        print(row)
        print('no :', row[0])
        print('title :', row[1])
        print('year :', row[2])
        print('score :', row[3])
        
        self.con.commit()           
```

```python
import csv # 파이썬 SQLite에 csv 파일 DB에 쓰기
import sqlite3
from sqlite3 import Error

def connection(): # db 연결
    try:
        con = sqlite3.connect('teamproject.db', timeout=10) # 안걸면 여러 개 진행이 어려움
        cursor_db = con.cursor()
        return con, cursor_db
    except Error:
        print(Error)

def create_table(cursor_db): # 테이블 생성
    cursor_db.execute("CREATE TABLE checkup(id integer PRIMARY KEY, Name text, Height real, Weight real, Measured_date text)")

def wr_csv(data): # csv 파일 작성
    with open('data.csv', 'w', newline='') as file:
        writer = csv.writer(file)
        writer.writerows(data)    

def csv_db(cursor_db): # csv 파일 읽어오기
    student = csv.reader(open('data.csv', 'r'), delimiter=',', quotechar='"')
    index = 0
    for row in student: # 테이블에 읽어온 값 입력하기
        index = index + 1
        print("%s %s" % (index, row))
        cursor_db.execute("INSERT INTO checkup VALUES (?, ?, ?, ?, ?)", (index, row[0], row[1], row[2], row[3]))

def ex_close(con): # sqlite 종료
    con.commit()
    con.close()

data = (['John','180cm','75kg','2022-09-08'],['Mina','179cm','101kg','2022-09-08'],['Dan','186cm','70kg','2022-09-08'],['Hart','172cm','65kg','2022-09-08'])

wr_csv(data)
con, cursor_db = connection()
create_table(cursor_db)
csv_db(cursor_db)
ex_close(con)
```
