```python
#크롤링 
from bs4 import BeautifulSoup
from selenium import webdriver
import time
import sys #저장 시 사용하는 모듈
import pandas as pd
import re
import requests
```

```python
#크롬 드라이버를 사용해서 웹 브라우저 실행
path = "C:/Users/hyunj/Downloads/chromedriver_win32/chromedriver.exe"
driver = webdriver.Chrome(path)

driver.get("https://www.mois.go.kr/frt/sub/a04/localGovernment/screen.do")
time.sleep(1) #인터넷 창이 모두 열릴 때까지 2초 기다림

gov_df = []
df = pd.DataFrame(gov_df, columns = ['title','hrefs', 'texts_1','text_urls_1', 'texts_2','text_urls_2', 'texts_3','text_urls_3'])

html = driver.page_source  #현재 화면의 html을 모두 불러옴
soup = BeautifulSoup(html, 'html.parser')  #불러오는 정보를 걸러주는 도구

list_titles = soup.select("print_area")
for j in range(16):
    for i in soup.find_all("ul", class_="location_list")[j].find_all("li"):
        title = i.get_text()
        hrefs = i.find("a")["href"]
        df = df.append({'title':title, 'hrefs':hrefs}, ignore_index=True)
df = df.append({'title':'세종특별자치시', 'hrefs':'https://www.sejong.go.kr/'}, ignore_index=True)
```

```python
from urllib.parse import quote_plus
for i in range(len(df['title'])):
    baseUrl = 'https://www.google.com/search?q='
    plusUrl = f'"{df["title"][i]}" {df["hrefs"][i]} 재활용품 배출 안내'
    url = baseUrl + quote_plus(plusUrl) # 한글을 변환해주는 변수

    driver = webdriver.Chrome()
    driver.get(url)

    html = driver.page_source
    soup = BeautifulSoup(html)

    v = soup.select('.yuRUbf')
    df['texts_1'][i] = v[0].select_one('.LC20lb.DKV0Md')# 검색결과 1 저장
    df['text_urls_1'][i] = v[0].a.attrs['href']

    df['texts_2'][i] = v[1].select_one('.LC20lb.DKV0Md')# 검색결과 2 저장
    df['text_urls_2'][i] = v[1].a.attrs['href']
    
    df['texts_3'][i] = v[2].select_one('.LC20lb.DKV0Md')# 3
    df['text_urls_3'][i] = v[2].a.attrs['href']
    
driver.close()
```
