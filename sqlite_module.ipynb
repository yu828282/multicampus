```python
class Aihub: # 모듈화
    def __init__(self): # 연결
        self.con = sqlite3.connect("teamproject.db", timeout=15, isolation_level=None)
        self.cur = self.con.cursor()
        
    def __del__(self): # 연결해제
        self.cur.close()
        self.con.close()
        
    def aihub_insert(self, file_name, label, update_date, x, y, box_x, box_y):
        sql = "INSERT INTO data_aihub ('file_name', 'label', 'update_date', 'x', 'y', 'box_x', 'box_y') VALUES (?,?,?,?,?,?,?)"
        self.con.execute(sql,(file_name, label, time.strftime('%Y-%m-%d'), x, y, box_x, box_y,))
        self.con.commit()
        
    def aihub_update(self, file_name, label, x, y, box_x, box_y, data_id):  
        sql = "UPDATE  data_aihub SET file_name = ?, label= ?, x =?, y =?, box_x =?, box_y =? where data_id =?"
        self.con.execute(sql,(file_name, label, x, y, box_x, box_y, data_id,))
        self.con.commit()
        
    def aihub_delete(self, data_id):             
        sql = "DELETE FROM data_aihub where data_id =?"
        self.con.execute(sql, (data_id,))
        self.con.commit()           
    
    def aihub_selectAll(self):            
        sql = "SELECT * FROM data_aihub"
        res = self.cur.execute(sql,)
        for _ in res:            
            print(_)
        return res
        
    def aihub_selectOne(self, data_id):      
        sql = "SELECT * FROM data_aihub Where  data_id = ?"
        self.cur.execute(sql, (data_id,))
        row = self.cur.fetchone()
        print(row)
        
        self.con.commit()           
```
